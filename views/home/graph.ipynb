```python
# !pip install numpy (colab이면 pass)
# !pip install pandas (colab이면 pass)
import numpy as np
import pandas as pd
```

```python
# !pip install matplotlib (colab이면 pass)
import matplotlib.pyplot as plt 
```

```python
x = np.arange(1000) # 0 ~ 1000 까지 numpy array 만들어줌
y = np.random.rand(1000)-0.5 # 1000개의 0~1까지 랜덤 , -0.5 해줘서 -0.5 ~ 0.5 까지로 변경
```

```python
plt.plot(x,y.cumsum())
# cumsum (누적합) -> 랜덤값 -0.5 ~ 0.5가 누적해서 계속 더해짐 
# 그래프 plt.plot(x,y)
```

```python
df = pd.DataFrame(y.cumsum(),index=x)
# Dataframe으로 변경
```

```python
df
```

```python

plot = df.iplot(asFigure=True, xTitle='Date', yTitle='Won')
# iplot 사용해서 Plotly 내부 객체로 지정 
# to_json() 하면 json형식으로 나옴 
# 이거 복사해서 coin.ejs에 써놓은 곳에 붙였음
# 파일로 만들어서 저장하고 불러오는 식으로 변경하는게 좋을듯
plot.to_json()
```
